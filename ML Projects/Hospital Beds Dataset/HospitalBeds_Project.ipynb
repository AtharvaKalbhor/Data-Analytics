INDIA ANALYSIS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
yearly_total_beds,yearly_percent = [], []
df = pd.read_csv("hospital_beds_per_india_v1.csv")
#Cleaning Data
df = df[df['state'] != 'IN']
df = df.drop(['country','county','measure','source','source_url','type','lat','lng','state'],axis = 1)
df = df.sort_values(by = "year")
year_col = df['year'].unique()
df.set_index("year",inplace = True)
#Total Bed Count
for year in year_col:
    bed_lst = df.loc[year]
    yearly_total_beds.append(((bed_lst["beds"]/1000)*bed_lst["population"]).sum())
df_yearly_total = pd.DataFrame(yearly_total_beds)

fig= plt.figure(figsize=(6,6))
axes= fig.add_axes([0.1,0.1,0.8,0.8])
plt.title("Yearwise Analysis of Total Bed Count of INDIA")
yearly_total_beds = [round(element,1) for element in yearly_total_beds]
yearly_percent = [round((element*100/sum(yearly_total_beds)),1) for element in yearly_total_beds]
axes.pie(yearly_total_beds,labels=year_col, autopct='%1.1f%%', startangle=90,counterclock = False,wedgeprops = {'linewidth': 10})
axes.legend(loc='lower right')
plt.show()

dict_data = dict(zip(yearly_total_beds,year_col))
yearly_percent.sort(reverse =True)
df_yearly = pd.DataFrame(yearly_total_beds,index = year_col,columns = ["TOTAL BED COUNT"]).T

print("Yearwise Bed Count of INDIA")
display(df_yearly)
print("""
\t\t\t\t\tINTERPRETATION
1.The Average Bed Count Of INDIA is {:.1f} and the Standard Deviation of Yearly Bed Count from 
  Average Bed Count is {:.1f}""".format(float(df_yearly_total.mean()),float(df_yearly_total.std()))) 
print("2.Total Number of Beds Accounted in India till 2018 are {:.2f}".format(sum(yearly_total_beds)))
print("3.The Most Number of beds Accounted in a Year were {:.2f}".format(max(yearly_total_beds)),"in the Year",dict_data[max(yearly_total_beds)],"which makes up",yearly_percent[0],"%")
print("4.The Least Number of beds Accounted in a Year were {:.2f}".format(min(yearly_total_beds)),"in the Year",dict_data[min(yearly_total_beds)],"which makes up",yearly_percent[-1],"%")
print("""
From the Above Analysis, 
  We may conclude that Most of the Analysis in INDIA was carried out in the Years 2015,2017,2018 which accounts 
to more than 90% of the Analysis.""")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
statewise_total_beds,percent_beds = [], []
df = pd.read_csv("hospital_beds_per_india_v1.csv")

#CLeaning Data
df = df.drop(['country','county','measure','source','source_url','type','lat','lng'],axis = 1)
df = df[df.state != 'IN']
df = df.sort_values(by = "state")
state_lst = df['state'].unique()
df.set_index("state",inplace = True)

#Fetching Statewise Bed Count
for state in state_lst:
    per_state_data = df.loc[state]
    statewise_total_beds.append((per_state_data["beds"]/1000)*per_state_data["population"])

fig= plt.figure(figsize=(10,6.5))
axes= fig.add_axes([0.1,0.1,0.8,0.8])
plt.title("Statewise Analysis Of Total Bed Count of INDIA")
plt.xlabel("States Of India")
plt.ylabel("Total Number of Beds")
axes.bar(state_lst,statewise_total_beds,color = ['orchid'], label="State")
xmin, xmax = plt.xlim()
plt.xlim(xmin*0.3, xmax *0.955)
plt.legend(loc="upper left")
plt.show()

total_beds_in_IND = sum(statewise_total_beds)
df_state_beds = pd.DataFrame(statewise_total_beds)
dict_data = dict(zip(statewise_total_beds,state_lst))
for state_total in statewise_total_beds:
    percent_beds.append((state_total/total_beds_in_IND)*100)

percentage_beds = [str(round(elem, 2))+"%" for elem in percent_beds ]
state_total = [str(round((elem/1000), 2))+"K" for elem in statewise_total_beds ]
df_percent_per_state = pd.DataFrame(list(zip(state_total,percentage_beds)),columns =["TOTAL BEDS","PERCENTAGE"],index = state_lst).T
pd.set_option('display.max_columns', 36)
print("\nStatewise Percentage of Bed Count in reference to Total Bed Count Of India")
display(df_percent_per_state)

print("\n\t\t\t\tINTERPRETATION")
print("1.The Total Number of Beds in INDIA : {:.1f} ".format(total_beds_in_IND))
print("2.The Average Number of Beds per State in INDIA : {:.1f} ".format(float(df_state_beds.mean())))
print("3.The Standard Deviation of Number of Beds per State from Average Number : {:.1f}".format(float(df_state_beds.std())))
print("4.Maximum Number of Beds in a State: {:.1f}".format(float(max(statewise_total_beds))),"in the State {} making up {:.1f}%".format(dict_data[max(statewise_total_beds)],float(sorted(percent_beds)[-1])))
print("5.Minimum Number of Beds in a State: {:.1f}".format(float(min(statewise_total_beds))),"in the State {} making up {:.1f}%".format(dict_data[min(statewise_total_beds)],float(sorted(percent_beds)[0])))
print("6.Number of States having Bed Count below Average : {}".format(int(df_state_beds[df_state_beds<df_state_beds.mean()].count())))
print("7.Number of States having Bed Count above Average : {}\n".format(int(df_state_beds[df_state_beds>df_state_beds.mean()].count())))
print("""From the Above Analysis,
  We may conclude that, the Number of Beds in India are non-uniformly distributed in the States as per the Standard Deviation.
States like Uttar Pradesh, Tamil Nadu, West Bengal and Karnataka have a Higher Bed Count, while Andaman and Nicobar Islands, 
Lakswadeep Islands and Daman and Deu have very low Count Of Beds.
  """)
```

USA ANALYSIS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
beds_per_type = []
df = pd.read_csv("hospital_beds_USA_v1.csv")
df = df.drop(['country','county','measure','source','source_url','lat','lng'],axis = 1)
df = df.sort_values(by = "type")
bed_type_lst = df['type'].unique() # Storing all types of beds available 
df.set_index("type",inplace= True)
for bed_type in bed_type_lst:
    bed_lst = df.loc[bed_type]
    beds_per_type.append(bed_lst["beds"].sum())   
beds_per_type = [round(element,1) for element in beds_per_type]    
df_beds_per_type = pd.DataFrame(beds_per_type)

plt.bar(bed_type_lst,beds_per_type, color ="crimson", label = 'Type of Bed')
plt.title("Types Of Beds Available in Hospitals of USA (2012-2020)")
plt.xlabel("Types Of Bed Available")
plt.ylabel("Total Number of Beds")
plt.legend(loc = 'upper right')
plt.show()
percent_per_bedtype = []
for bed in beds_per_type:
    percent_per_bedtype.append((bed/sum(beds_per_type))*100)
percent_per_bedtype = [(round(element,2)) for element in percent_per_bedtype]

percentage_per_bedtype = [str(element)+'%' for element in percent_per_bedtype]
df1 = pd.DataFrame(list(zip(beds_per_type,percentage_per_bedtype)),index = bed_type_lst, columns = ["TOTAL BED COUNT","PERCENTAGE"]).T
display(df1)

print("""\t\t\t\tINTERPRETATION
 
1.Maximum Amount of Beds are allocated to the Acute (short term) Diseases (ACUTE) which account to {}%.
2.They are followed by Chronic (long term) Diseases (ICU) which account to {}%.
3.Minimum Amount of Beds are allocated to OTHERS which account to {}%""".format(sorted(percent_per_bedtype)[-1],sorted(percent_per_bedtype)[-2],sorted(percent_per_bedtype)[0]))
print("3.Average Bed Count Per Type Of Beds : {:.1f}".format(sum(beds_per_type)/len(beds_per_type)))
print("4. Standard Deviation per Bedtype : {:.1f}".format(float(df_beds_per_type.std())))
print("""
From the Above Analysis,
  We may conclude that most of the focus is given to Acute and ICU beds while very few attention is given
to Psychiatric beds.
""")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
yearly_total_beds = []

#CLeaning
df = pd.read_csv("hospital_beds_USA_v1.csv")
df = df.drop(['country','county','measure','source','source_url','type','lat','lng','state'],axis = 1)
df = df[df.beds != 0]
df = df.sort_values(by = "year")
year_col = df['year'].unique()
df.set_index("year",inplace = True)

for year in year_col:
    bed_lst = df.loc[year]
    yearly_total_beds.append(((bed_lst["beds"]/1000)*bed_lst["population"]).sum())
df_yearly_bedcount = pd.DataFrame(yearly_total_beds)
yearly_percent = [round((element*100/sum(yearly_total_beds)),1) for element in yearly_total_beds]

#PLotting the Graph
fig= plt.figure(figsize=(9,6))
plt.plot(year_col,yearly_total_beds,label = "Total Number of Beds Added per Year",linewidth = 3,marker = "o",markeredgecolor = "orchid",markerfacecolor = "crimson")
for x_cor,y_cor in zip(year_col,yearly_total_beds):
    plt.text(x_cor,y_cor,'({:.2f})'.format(y_cor),horizontalalignment='left', verticalalignment='bottom')
plt.title("Yearwise Analysis Of Number of Beds Added in USA")
plt.xlabel("Year")
plt.ylabel('Total Number Of Beds')
plt.legend(loc = "upper left")
plt.show()

yearly_total = [str(round((element/1000),2))+"k" for element in yearly_total_beds]
percent_beds = [str(round(((element/sum(yearly_total_beds))*100),2))+"%" for element in yearly_total_beds]
df_yearly = pd.DataFrame(list(zip(yearly_total,percent_beds)),index = year_col,columns = ["TOTAL BED COUNT","PERCENTAGE"]).T
print("\t\t\tYearwise Analysis")
display(df_yearly)

dict_data = dict(zip(yearly_total_beds,year_col))
print("\n\t\t\t\tINTERPRETATION")
print("1.Total Number of Beds Accounted in USA till 2020 are {:.2f}".format(sum(yearly_total_beds)))
print("""2.The Average Yearly Bed Count of USA is {:.1f}
3.the Standard Deviation of Yearly Bed Count from Average Bed Count of USA is {:.1f}  """.format(float(df_yearly_bedcount.mean()),float(df_yearly_bedcount.std())))
print("4.The Most Number of beds Accounted in a Year were {:.2f}".format(max(yearly_total_beds)),"in the Year",dict_data[max(yearly_total_beds)],"which makes up {:.1f}%".format(sorted(yearly_percent)[-1]))
print("5.It is followed by {:.2f} beds in the Year {} which makes upto {}%".format(sorted(yearly_total_beds)[-2],dict_data[sorted(yearly_total_beds)[-2]],sorted(yearly_percent)[-2]))
print("6.The Least Number of beds Accounted were {:.2f}".format(min(yearly_total_beds)),"in the Year",dict_data[min(yearly_total_beds)],"which makes up {}%".format(sorted(yearly_percent)[0]))
print("""
From the above Analysis,
  We may conclude that in 2017 and 2018, USA had the maximum growth rate in Number of Beds as compared to the Rest of the Years.
In the other years, the total bed count is almost negligible as compared to Year 2018, due to which we see a huge Deviation from 
the Average Yearly Bed Count.""")
```

INDIA VS USA ANALYSIS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
yearly_total_ind,yearly_total_usa = [], []
#INDIA
#Cleaning
df_ind = pd.read_csv("hospital_beds_per_india_v1.csv")
df_ind = df_ind[df_ind['state'] != 'IN']
df_ind = df_ind.drop(['country','county','measure','source','source_url','type','lat','lng','state'],axis = 1)
df_ind = df_ind.sort_values(by = "year")
year_col_ind = df_ind['year'].unique()
df_ind.set_index("year",inplace = True)

for year in year_col_ind:
    bed_lst = df_ind.loc[year]
    yearly_total_ind.append(((bed_lst["beds"]/1000)*bed_lst["population"]).sum())
yearly_total_ind = [round(element,1) for element in yearly_total_ind]

#Plotting the Bar Graph for INDIA
fig= plt.figure(figsize=(10,7))
axes= fig.add_axes([0.1,0.1,0.8,0.8])
plt.xticks(range(2012,2021))
plt.title("Yearwise Analysis Of Total Bed Count of INDIA vs USA")
plt.xlabel("Year")
plt.ylabel("Total Number of Beds")
ind_bar = axes.bar(year_col_ind+0.2,yearly_total_ind,width = 0.4,label= "IND")

#USA
#CLeaning
df_usa = pd.read_csv("hospital_beds_USA_v1.csv")
df_usa = df_usa.drop(['country','county','measure','source','source_url','type','lat','lng','state'],axis = 1)
df_usa = df_usa[df_usa.beds != 0]
df_usa = df_usa.sort_values(by = "year")
year_col_usa = df_usa['year'].unique()
df_usa.set_index("year",inplace = True)

for year in year_col_usa:
    bed_lst = df_usa.loc[year]
    yearly_total_usa.append(((bed_lst["beds"]/1000)*bed_lst["population"]).sum())
yearly_total_usa = [round(element,1) for element in yearly_total_usa]

#Plotting the Bar Graph for USA
usa_bar = axes.bar(year_col_usa-0.2,yearly_total_usa,width=0.4,label = "USA")
x = np.arange(len(year_col_usa))

#To label the Bar Graphs
def autolabel(rects,x):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        axes.annotate('{:.1f}K'.format(height/1000),xy=(rect.get_x() + rect.get_width() / 2, height),xytext=(x, 3),textcoords="offset points",ha='center', va='bottom')
autolabel(ind_bar,3)
autolabel(usa_bar,-3)
plt.legend(loc='upper left')
plt.show()

#DataFrame for Analysis
print("Percentage of Yearwise Bed Count to the Total Bed Count of 2012-2020 for INDIA and USA")
percent_ind = [str(round(((element/sum(yearly_total_ind))*100),2))+"%" for element in yearly_total_ind]
percent_usa = [str(round(((element/sum(yearly_total_usa))*100),2))+"%" for element in yearly_total_usa]
percent_ind = ["0.00%"]*2 + percent_ind + ["0.00%"]*2
df_percent = pd.DataFrame(list(zip(percent_ind,percent_usa)),columns = ["INDIA","USA"],index = range(2012,2021)).T
display(df_percent)

print("""\t\t\t\t\tANALYSIS
1.The Yearwise Analysis of India was during 2014 to 2018, while that of USA was from 2012 to 2020.
2.As for 2020, USA has a Total of {:.1f} beds, while INDIA has a tally of {:.1f} beds.""".format(sum(yearly_total_usa),sum(yearly_total_ind)))
print("""3.INDIA dominated  USA in Total Bed Count in the Year {} and {} by a big margin of {} beds 
  and {} beds respectively.""".format(year_col_usa[3],year_col_usa[5],(yearly_total_ind[1]-yearly_total_usa[3]),
                                      (yearly_total_ind[3]-yearly_total_usa[5])))
print("4.In the Year {},{}, USA had a narrow lead of {} and {} beds respectively.".format(year_col_usa[2],year_col_usa[4],
                                                                                          (yearly_total_usa[2]-yearly_total_ind[0]),(yearly_total_usa[4]-yearly_total_ind[2])))
print("""5.USA was behind India in Total Bed Count till 2017, having a Total Bed Count of {} beds 
  while INDIA had {} beds.""".format(sum(yearly_total_usa[:6]),sum(yearly_total_ind[:4]))) 
print("""6.It took a huge leap in the Year {}, where the Total Bed Count of that year was a whopping {} beds 
  as opposed to INDIA's {} beds.""".format(year_col_usa[6],yearly_total_usa[6],yearly_total_ind[-1])) 

df_yearly_beds_usa = pd.DataFrame(yearly_total_usa)
df_yearly_beds_ind = pd.DataFrame(yearly_total_ind)

print("""
\t\t\t\t\tINTERPRETATION
1. The Average Bed Count Of INDIA is {:.1f} and the Standard Deviation of Yearly Bed Count from 
  Average Bed Count is {:.1f} 
2. The Average Bed Count Of USA is {:.1f} and the Standard Deviation of Yearly Bed Count from 
  Average Bed Count is {:.1f} 
""".format(float(df_yearly_beds_ind.mean()),float(df_yearly_beds_ind.std()),float(df_yearly_beds_usa.mean()),float(df_yearly_beds_usa.std())))
print("""
From the Above Analysis, 
  We may conclude that the Analysis carried out in INDIA was more evenly carried out than it was in USA, 
as shown by the Standard Deviation. Also, the Year 2017 was crucial for the Analysis in INDIA while 
Year 2018 was crucial for USA, as both the countries had the maximum growth rate in Count of Number 
of Hospital Beds in the respective year.""")
```
